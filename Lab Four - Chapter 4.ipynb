```r
# libraries and necessities

library(ggplot2)
```

```r
# import and interrogate the University data

schools.df <- read.csv("Datasets/Universities.csv",header = TRUE)

dim(schools.df)
t(t(names(schools.df)))
head(schools.df)
summary(schools.df)
colnames(schools.df)
```

```r
# Remove all categorical variables (although I'm keeping the school name to act as a key)

schoolNoFacts.df <- within(schools.df, rm(State, "Public..1...Private..2."))

dim(schoolNoFacts.df)
```

```r
# Remove all records with missing numerical measurements from the dataset

schoolNoFactsNoNA.df <- schoolNoFacts.df

#  I found this delicious function on the googles
#   ...basically, if any value on the row is NA, then drop the row

schoolNoFactsNoNA.df <- schoolNoFactsNoNA.df[complete.cases(schoolNoFactsNoNA.df),]
```

```r
# Attempt to run the PCA on every column in the dataframe except the first

pca <- prcomp(schoolNoFactsNoNA.df[,-1])

summary(pca)
pca$rot[,1:3]
```
