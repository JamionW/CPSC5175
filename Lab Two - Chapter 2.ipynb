### Problem 2.6
New data bought and fed into a prospecting model shows that "refund issued" is a useful predictor. Why is this not appropriate for model inclusion?

#### Answer
An example of not being able to see the forest for the trees, it is not appropriate to use variables which directly depend on each other as a function of their business, such as "Tax Paid": it will always be a representation of a real-life percentage of price and is a byproduct of the variable you are trying to predict, not a predecessor. 

Similarly, the price of an item _begets_ the amount refunded, not the other way around.

### Problem 2.11b
Prepare and partition the Toyota Corolla dataset for modeling.

#### Answer

```r
#data import and basic metrics

corolla.df <- read.csv("Datasets/ToyotaCorolla.csv",header = TRUE)

dim(corolla.df)
t(t(names(corolla.df)))
head(corolla.df)
```

```r
# ...of the above, the two good factor columns that we can use to dummy out are Model and Fuel Type. 
#  Let's look at their distribution:

factor(corolla.df$Model)
```

```r
# well, that's no good. Let's try fuel:

factor(corolla.df$Fuel)
```

```r
# Much better. Now we can turn these into dummies.

dummyframe <- model.matrix(~ corolla.df$Fuel) 
dummyframe <- as.data.frame(dummyframe)

t(t(names(dummyframe)))

dummyframe <- dummyframe [,-1]

head(dummyframe)
```

```r
# So now, let's concatenate those into the dataframe:

corolla.df <- cbind(corolla.df, dummyframe)

t(t(names(corolla.df)))
```

```r
# Now let's partition that into training, validation, and holdout sets

set.seed(1)

training.rows <- sample(rownames(corolla.df), dim(corolla.df)[1]*0.4)
training.data <- corolla.df[training.rows,]

validation.rows <- sample(setdiff(rownames(corolla.df),training.rows), dim(corolla.df)[1]*0.3)
validation.data <- corolla.df[validation.rows,]

holdout.rows <- setdiff(rownames(corolla.df), union(training.rows, validation.rows))
holdout.data <- corolla.df[holdout.rows,]

print("Training:")
dim(training.data)

print("Validation:")
dim(validation.data)

print("Holdout:")
dim(holdout.data)
```

##### Conclusion

For the above, the Training dataset will be used to build the model and the validation dataset will be used to validate the effectiveness of our model. For further confirmation, the holdout dataset will be used to feed fresh data into our model for further analysis and verification.
