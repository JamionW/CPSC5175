### Problem 3.2
##### Sale of riding mowers: Scatter Plots

```r
# libraries and necessities

library(ggplot2)
options(repr.plot.width=10, repr.plot.height=5) 

# import and interrogate the Mower data

mowers.df <- read.csv("Datasets/RidingMowers.csv",header = TRUE)

dim(mowers.df)
t(t(names(mowers.df)))
head(mowers.df)

levels(mowers.df$Ownership)
```

```r
# attempt the scatter plot

# needed to format the width and height into something sensible
options(plot.width=7, plot.height=4)

ggplot(mowers.df, aes(x = Lot_Size, y = Income)) +
    # establish the key on the factor 
  geom_point(size = 1, aes(colour = factor(Ownership))) +
    # name the axis and title
    labs(title = "Prospective Owners", 
         subtitle = "Income compared to Lot Size",
         colour = "Ownership") +
    xlab("Lot Size")
```

### Problem 3.3
##### Laptop Sales at a London Computer Chain: Bar Charts and Boxplots

```r
# libraries and necessities

library(ggplot2)
library(tidyverse)
library(scales)

# import and interrogate the Mower data

laptops.df <- read.csv("Datasets/LaptopSalesJanuary2008.csv",header = TRUE)

# this is me creating an arbitrary "store number" out of the X+Y coordinates
# it ain't pretty

laptops.df$Store <- paste0(str_trim(laptops.df$OS.X.Store),".",str_trim(laptops.df$OS.Y.Store))
laptops.df$totals <- as.integer(laptops.df$Retail.Price)

dim(laptops.df)
t(t(names(laptops.df)))
head(laptops.df)

summary(laptops.df)
```

```r
# For Part A: Bar Chart

# create aggregate dataframe per store

aggplot <- aggregate(laptops.df$totals, by = list(laptops.df$Store), FUN = mean)
names(aggplot) <- c("Store","Average.Price")
```

```r
# and then make our plot

options(plot.width=7, plot.height=4)

aggplot %>%
    ggplot(aes(reorder(Store, Average.Price), Average.Price)) + 
    geom_col() + 
    coord_flip() +
    # I realize this is dumb but don't want to touch it
    scale_y_continuous(label=comma)
```

```r
# For Part B: Boxplot

# knowing now from the above that store 533788.189994 has the highest average and
#  store 519585.177640 has the lowest, we can compare the distributions in the original 
#  dataframe by going against just those two stores

twostores <- laptops.df[which(laptops.df$Store=='533788.189994' | laptops.df$Store=='519585.177640'),]

summary(twostores)
```

```r
# Now, run the boxplot for just those two stores

twostores %>%
    ggplot(aes(Store, totals)) +
    geom_boxplot()

# and, surprise! The higher distribution of prices per unit result in a vast aggregate difference
```
