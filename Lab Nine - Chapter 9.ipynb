### Problem 9.1
##### Work upon the eBayAuctions.csv dataset.

```r
# necessities

library(caret)
library(reshape2)
library(rpart)
library(rpart.plot)
set.seed(420)
```

```r
# Import the dataset and interrogate it.

raw.df <- read.csv("Datasets/eBayAuctions.csv",header = TRUE)

str(raw.df)

unique(raw.df$Duration)
```

```r
# split out Duration as a factor and add an ID

pre.df <- raw.df[,-c(4)]

pre.df$Duration <- factor(raw.df$Duration)

pre.df$ID <- seq.int(nrow(raw.df))

str(pre.df)
```

```r
# split into train and test

set.seed(420)

trainList <- createDataPartition(pre.df$ID, p = 0.6, list = FALSE, times = 1)
train.df <- pre.df[trainList,]
test.df <- pre.df[trainList,]
```

### Problem 9.1a
##### split into a best-pruned tree

```r
set.seed(420)

# specifications per the book's problem

rpartFit <- rpart(Duration ~., data=train.df, control=rpart.control(maxdepth=7, minsplit=50), method="class")
```

```r
# I messed around with the options because I thought simplifying the inner nodes (type=5) made more sense

prp(rpartFit, type=5, extra=+100, split.font=2, varlen=0, box.palette="auto")
```
